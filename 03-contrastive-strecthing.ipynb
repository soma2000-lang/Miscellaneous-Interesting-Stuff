```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../images/low-contrast.png', cv2.IMREAD_GRAYSCALE)

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

r_min, r_max = np.min(image), np.max(image)

# Apply contrast stretching
# s = (r - r_min) * ((L-1) / (r_max - r_min))
stretched_image = ((image - r_min) / (r_max - r_min) * 255).astype(np.uint8)

plt.subplot(1, 3, 2)
plt.title("Stretched Image")
plt.imshow(stretched_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Histograms")
plt.hist(image.ravel(), bins=256, range=(0, 255), color='blue', alpha=0.5, label='Original')
plt.hist(stretched_image.ravel(), bins=256, range=(0, 255), color='red', alpha=0.5, label='Stretched')
plt.legend(loc="upper right")

plt.tight_layout()
plt.show()
```
